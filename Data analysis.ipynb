```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('leads_basic_details.csv')
df2=pd.read_csv('leads_demo_watched_details.csv')
df3=pd.read_csv('leads_interaction_details.csv')
df4=pd.read_csv('leads_reasons_for_no_interest.csv')
df5=pd.read_csv('sales_managers_assigned_leads_details.csv')
```

```python
df1
```

```python
sns.boxplot(df1['age'])
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1

df1 = df1[~((df1 < (Q1 - 1.5 * IQR)) |(df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
sns.boxplot(df1['age'])
```

## Most role of parents occupation

```python
ax=df1['parent_occupation'].value_counts().plot(kind='bar',color='b')
ax.tick_params(axis='x', labelrotation=45)
```

â€¢ Most parents are are government employees and least are in educational field

# Lead's Educational background

```python
ax=df1['current_education'].value_counts().plot(kind='bar',color='orange')
ax.tick_params(axis='x', labelrotation=45)
```

# Most lead sources

```python
ax=df1['lead_gen_source'].value_counts().plot(kind='pie',autopct='%1.0f%%',radius=1.5,colors = ['aqua', 'grey','red','yellow','b'])
plt.title('Lead Sources', pad=32)
ax.yaxis.set_visible(False)
```

â€¢ Majority of leads are from social media and least leads are from referrals

# Most residing cities 

```python
fig_dims = (9,5)
fig, ax = plt.subplots(figsize=fig_dims)

ax=df1['current_city'].value_counts().plot(kind='bar',color='r')
ax.tick_params(axis='x', labelrotation=0)
```

```python
ax=df2['language'].value_counts().plot(kind='bar',color='green')
ax.tick_params(axis='x', labelrotation=0)
```

â€¢ Many people prefered watching the demo in english and least amount of people watched it in hindi and a moderate amount of people watched the demo in Telgu because we have many leads residing in Hyderabad & Vishakapatnam

```python
ax=df3['call_reason'].value_counts().plot(kind='bar',color='lime')
ax.tick_params(axis='x', labelrotation=45)
```

â€¢ Here we can see majority of leads did not attend the demo so in order to get follow up a call was made

# Reasons for not interested in demo

```python
g=df4.groupby('reasons_for_not_interested_in_demo').size().plot(kind='pie',autopct='%1.0f%%',radius=1.5,colors = ['darkorange', 'khaki','yellow','darkgoldenrod','yellowgreen'])
plt.title('Lead Sources', pad=32)
```

# Reasons for not interested to consider

```python
g=df4.groupby('reasons_for_not_interested_to_consider').size().plot(kind='pie',autopct='%1.0f%%',radius=1.5,colors = ['darkorange', 'khaki','yellow','darkgoldenrod','yellowgreen'])
plt.title('Lead Sources', pad=50)
```

# Reasons for not interested to convert

```python
g=df4.groupby('reasons_for_not_interested_to_convert').size().plot(kind='pie',autopct='%1.0f%%',radius=1.5,colors = ['darkorange', 'khaki','yellow','darkgoldenrod','yellowgreen'])
plt.title('Lead Sources', pad=40)
```
